### 1

Usando la ecuación de Nernst, determinar los potenciales de equilibrio para los
siguientes iones: $K$, $N$ a y $Cl$. Las concentraciones son:

|         | Interior (mM) | Exterior (mM)  |
|:-------:|:-------------:|:--------------:|
| $K^+$   | 430           | 20             |
| $Na^+$  | 50            | 440            |
| $Cl^-$  | 65            | 550            |

La temperatura es de 20 grados centígrados. Ayuda: el valor de la constante de
Boltzmann es de $K_B ≈ 8,6 10^{−5}eV/K$.


#### Solución

La ecuación de Nernst

$$\Delta V = \frac{K_B T}{Z e} ln(\frac{C_{out}}{C_{in}}) $$

nos permite obtener los potenciales ($\Delta V$) de equilibrio para cada ion, donde:

+ $Z$ valencia del ion
+ $e$ carga del electrón
+ $T$ temperatura
+ $K_B$ constante de Boltzmann
+ $C_{in}$ concentración interior
+ $C_{out}$ concentración exterior

```python
from math import log
data = [("$K^+ $",+1, 430., 20. ),
        ("$Na^+$",+1, 50. , 440.),
        ("$Cl^-$",-1, 65. , 550.)]

T, KB = 20+273.15, 8.6e-5

# No es necesario dividir por e porque [eV]/e = [V]
print_table([[""  ,"$\Delta V$"]]+\
            [[name,"${:.3} mV$".format(1e3* (KB*T)/Z*log(Cout/Cin)) ] for name, Z, Cin, Cout in data])
```

### 2

Considerar una neurona esférica con un radio de 15 micrones y una capacitancia de 1 µF/cm2. ¿Qué cantidad de iones de Sodio debe ingresar a la neurona para cambiar el potencial de membrana en 100 mV? Comparar el cambio de concentración con la concentración de iones de Sodio del problema anterior. Ayuda: usar como valor de la
constante de Faraday F = 102 C/mol.

#### Solución

La capacidad de la neurona es $C = 1 µF/cm^2 \: 4\pi (15µm)^2 = 28.3 pF$

El cambio de carga en la neurona para que su potencial aumente en $100 mV$ es $\Delta q=100 mV \times 28.3 pF= 2.83 pC$

La cantidad de iones Sodio que deben ingresar es $n = \Delta q/e = 2.83 pC /1.6\times 10^{-19} C =\fbox{$1.77\times 10^7$}$
